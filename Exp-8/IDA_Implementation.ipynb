```python
# ----------------------------
# Initial Data Analysis (IDA) on Iris Dataset
# ----------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load dataset
data = pd.read_csv("IRIS.csv")

# Step 2: Basic info
print("First 5 rows of dataset:\n", data.head())
print("\nShape of dataset:", data.shape)
print("\nData types:\n", data.dtypes)
print("\nMissing values:\n", data.isnull().sum())

# Step 3: Summary statistics
print("\nSummary statistics:\n", data.describe())

# Step 4: Class distribution
print("\nClass distribution:\n", data['species'].value_counts())

# Step 5: Histograms of features
data.hist(figsize=(10,8))
plt.suptitle("Feature Distributions", fontsize=14)
plt.show()

# Step 6: Boxplots by species
plt.figure(figsize=(12,8))
for i, col in enumerate(data.columns[:-1]):
    plt.subplot(2,2,i+1)
    sns.boxplot(x="species", y=col, data=data)
plt.title("Boxplots of Features by Species")
plt.tight_layout()
plt.show()

# Step 7: Correlation heatmap
plt.figure(figsize=(6,4))
sns.heatmap(data.iloc[:,:-1].corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()

# Step 8: Pairplot
sns.pairplot(data, hue="species", diag_kind="kde")
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```
